```python
def fiyat_kategori(price):
    category = np.full_like(price , 'Ucuz' , dtype='U10')
    category[(price > 5000)  & (price < 12999)] = 'Orta'
    category[price > 13000] = 'Pahalı'
    return category
```

```python
df['Price_Category'] = fiyat_kategori(df['Price'])
```

```python
df.sample(10)
```

```python
df.drop('Price_Category' , axis=1 , inplace=True)
```

```python
def yil(year):
    kutu = np.full_like(year, 'Eski', dtype='U10')
    kutu[(year > 2005) & (year <= 2012)] = 'Orta'
    kutu[year >= 2013] = 'Yeni'
    return kutu
```

```python
df['Year_Category'] = yil(df['Year'])
```
